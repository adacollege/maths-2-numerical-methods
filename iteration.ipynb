Show that $x^3-3x^2-2x+5=0$ has a *root* in the *interval* $3<x<4$

Show that
$$x^3-3x^2-2x+5=0$$
can be rewritten as
$$x=\sqrt{\dfrac{x^3-2x+5}{3}}$$

Use desmos to plot the graphs of $y=x$ and $y=\sqrt{\dfrac{x^3-2x+5}{3}}$

Where do the two graphs intersect? What does that mean?

```python
xs = np.linspace(-2,6,100)
y1s = xs
y2s = np.sqrt((xs**3-2*xs+5)/3)

plt.plot(xs,y1s)
plt.plot(xs,y2s);
```

```python
xs = np.linspace(-2,4,100)
y1s = xs
y2s = (xs**3-3*xs**2+5)/2

plt.plot(xs,y1s)
plt.plot(xs,y2s);
```

```python
xs = np.linspace(-2,4,100)
y1s = xs
y2s = np.cbrt(3*xs**2+2*xs-5)

plt.plot(xs,y1s)
plt.plot(xs,y2s);
```

```python
def f(x):
    return (x**3-3*x**2+5)/2
```

```python
def g(x):
    return x**3-3*x**2-2*x+5
```

```python
def root_between(a,b):
    if g(a) < 0 and g(b) > 0:
        return True
    elif g(b) < 0 and g(a) < 0:
        return True
    else:
        return False
```

```python
root_between(3,4)
```

```python
root_between(1,2)
```

```python
def root_between(a,b):
    if g(a) * g(b) < 0:
        return True
    else:
        return False
```

```python
root_between(1,2)
```
