Show that
$$x^3-3x^2-2x+5=0$$
can be rewritten as
$$x=\sqrt{\dfrac{x^3-2x+5}{3}}$$

Use desmos to plot the graphs of $y=x$ and $y=\sqrt{\dfrac{x^3-2x+5}{3}}$

Where do the two graphs intersect? What does that mean?

```python
xs = np.linspace(-2,6,100)
y1s = xs
y2s = np.sqrt((xs**3-2*xs+5)/3)

plt.plot(xs,y1s)
plt.plot(xs,y2s);
```

```python
xs = np.linspace(-2,4,100)
y1s = xs
y2s = (xs**3-3*xs**2+5)/2

plt.plot(xs,y1s)
plt.plot(xs,y2s);
```

```python
xs = np.linspace(-2,4,100)
y1s = xs
y2s = np.cbrt(3*xs**2+2*xs-5)

plt.plot(xs,y1s)
plt.plot(xs,y2s);
```

```python
#the rearranged version
def f(x):
    return (x**3-3*x**2+2)/2
```

```python
#the oringinal function
def g(x):
    return x**3-3*x**2-2*x+5
```

```python
def root_between(a,b):
    if g(a)*g(b) > 0:
        return False
    else: 
        return True
    
#returning f(a) will show the y value when x=a 
#returning f(b) will show the y value when x=b
#if one of the y values is nagative and the other is positive then there is a solution of the function between a and b

    
```
