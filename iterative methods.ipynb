```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Numerical Methods

> Solve the equation $x=\cos{x}$

```python
fig, ax = plt.subplots()
ax.axison = False
xs = np.linspace(-0.5,2)
y1s = xs
y2s = np.cos(xs)
ax.axvline(0)
ax.axhline(0)
ax.plot(xs,y1s,'g',lw=5)
ax.plot(xs,y2s,'b',lw=5)
ax.annotate(xy=(0.74, 0.74), xytext=(0.5,1.3), s="$x=\cos{x}$", arrowprops={'arrowstyle':'->'});
```

### Interval bisection

Equivalently,

> Solve the equation $x-\cos{x}=0$

```python
fig, ax = plt.subplots()
ax.axison = False
xs = np.linspace(-0.5,2)
y1s = xs
y2s = np.cos(xs)
y3s = xs - np.cos(xs)
ax.axvline(0)
ax.axhline(0)
ax.plot(xs,y1s,'g',lw=5, alpha=0.2)
ax.plot(xs,y2s,'b',lw=5, alpha=0.2)
ax.plot(xs,y3s,'r',lw=5)
ax.annotate(xy=(0.74, 0.74), xytext=(0.5,1.3), s="$x=\cos{x}$", arrowprops={'arrowstyle':'->'})
ax.annotate(xy=(0.74, 0), xytext=(0.9,-0.5), s="$x-\cos{x}=0$", arrowprops={'arrowstyle':'->'});
```

When $x=0$, $x-\cos{x}=0-1=-1<0$

When $x=1$, $x-\cos{x}=1-\cos{1}=0.46>0$

Therefore, the equation has a solution in the interval $(0,1)$

```python
fig, ax = plt.subplots()
ax.axison = False
xs = np.linspace(-0.5,2)
y3s = xs - np.cos(xs)
ax.axvline(0)
ax.axhline(0)
ax.axvline(0,linestyle='--',color='k')
ax.axvline(1,linestyle='--',color='k')
ax.plot(xs,y3s,'r',lw=5)
ax.annotate(xy=(0, 0), s="$0$")
ax.annotate(xy=(1, 0), s="$1$");
```

When $x=0.5$, $x-\cos{x}=0.5-\cos{0.5}=-0.38<0$

So the solution is in the interval $(0.5,1)$

```python
fig, ax = plt.subplots()
ax.axison = False
xs = np.linspace(-0.5,2)
y3s = xs - np.cos(xs)
ax.axvline(0)
ax.axhline(0)
ax.axvline(0.5,linestyle='--',color='k')
ax.axvline(1,linestyle='--',color='k')
ax.plot(xs,y3s,'r',lw=5)
ax.annotate(xy=(0.5, 0), s="$0.5$")
ax.annotate(xy=(1, 0), s="$1$");
```

> Continue *bisecting* intervals until you know the solution to 2 decimal places

> How do you *know* that your solution is correct to 2 d.p.?

## Fixed-point iteration

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.axison = False
xs = np.linspace(-0.5,2)
y1s = xs
y2s = np.cos(xs)
ax.axvline(0)
ax.axhline(0)
ax.plot(xs,y1s,'g',lw=5, alpha=0.4)
ax.plot(xs,y2s,'b',lw=5, alpha=0.4)
xi = [2]
yi = []
for i in range(5):
    yi.append(np.cos(xi[-1]))
    xi.append(yi[-1])
    yi.append(yi[-1])
    xi.append(yi[-1])
ax.plot(xi[:-1],yi,lw=3)
for i,x in enumerate(xi[0::2]):
    ax.annotate(xy=(x,0),xytext=(x,0.1),s=f'$x_{i}=${round(x,2)}',rotation=90);
```

> Show that the equation
> $$x^3-x-1=0$$
> has a root, $\alpha$, in the interval $(0,2)$
>
> Use the rearrangement
> $$x=\sqrt[3]{x+1}$$
> to find the value of $\alpha$ correct to 2 d.p.,
> illustrating your method with a *cobweb diagram*.

```python
fig, ax = plt.subplots(figsize=(10,6))
fig.suptitle("$x_{n+1}=\sqrt[3]{x_n+1}$")
ax.axison = False
ax.set_ylim(-1,2)
xs = np.linspace(-0.5,2,100)
y1s = xs
y2s = (xs+1)**(1/3)
ax.axvline(0)
ax.axhline(0)
ax.plot(xs,y1s,'g',lw=5, alpha=0.4)
ax.plot(xs,y2s,'b',lw=5, alpha=0.4)
xi = [0]
yi = []
for i in range(5):
    yi.append((xi[-1]+1)**(1/3))
    xi.append(yi[-1])
    yi.append(yi[-1])
    xi.append(yi[-1])
ax.plot(xi[:-1],yi,lw=3);
```

What happens if instead we use the following rearrangement?$$x=x^3-1$$

```python
fig, ax = plt.subplots(figsize=(10,6))
fig.suptitle("$x_{n+1}=(x_{n+1})^3 - 1$")
ax.axison = False
ax.set_ylim(-5,3)
ax.set_xlim(-3,2)
xs = np.linspace(-3,2,100)
y1s = xs
y2s = (xs)**3 - 1
ax.axvline(0)
ax.axhline(0)
ax.plot(xs,y1s,'g',lw=5, alpha=0.4)
ax.plot(xs,y2s,'b',lw=5, alpha=0.4)
xi = [0]
yi = []
for i in range(3):
    yi.append(xi[-1]**3-1)
    xi.append(yi[-1])
    yi.append(yi[-1])
    xi.append(yi[-1])
ax.plot(xi[:-1],yi,lw=3);
```
